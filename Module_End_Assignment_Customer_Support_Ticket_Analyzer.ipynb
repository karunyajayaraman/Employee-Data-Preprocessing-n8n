```python
#Task: Print the initial ticket data in a readable format.

for i in range(len(ticket_data['Ticket_No'])):

#need the index (i) to access values from one or more columns
#i-to access the corresponding row in each column

    print(f"Ticket No        : {ticket_data['Ticket_No'][i]}")
    print(f"Customer Name    : {ticket_data['Customer_Name'][i]}")
    print(f"Issue            : {ticket_data['Issue_Description'][i]}")
    print(f"Priority         : {ticket_data['Priority'][i]}")
    print("-" * 70)


#Step 2: Add More Tickets

ticket_count = int(input("How many new tickets do you want to add?"))  # get the input of ticket from user(How many tickets?)

next_ticket = ticket_data['Ticket_No'][-1] + 1   # increamenting the ticket +1 (Generate next ticket number)

for i in range (ticket_count):
  print(f'\nEnter the Ticket Details {next_ticket}')
  cust_name = input("Customer Name : ")    #Get user input - customer name & Issue Description
  issue_desc = input("Issue Decription : ")

#check the priority - condition
  while True:
        priority = input("Priority (High/Medium/Low): ").title()   # Get user input(high,medium,low)

        if priority in ["High", "Medium", "Low"]:
            break
        else:
            print("Invalid Priority! Please enter High, Medium or Low.")

  ticket_data['Ticket_No'].append(next_ticket)   #Append all values to their respective lists
  ticket_data['Customer_Name'].append(cust_name)
  ticket_data['Issue_Description'].append(issue_desc)
  ticket_data['Priority'].append(priority)

  next_ticket += 1   #Increase ticket number
```

```python
#Step 3: Text Cleaning for Issue Descriptions

#Hints: Use .replace(), .split(), ' '.join(), .strip(), .lower().

cleaned_issues = [] # instead of stored in existing 'issue_description' create a new empty list to store

for issue in ticket_data['Issue_Description']:# cleaning the Issue description

  issue =  issue.lower()   # convert to lowercase

  #Remove punctuation (.,!?-')
  #symbol = ['.', ',', '!', '?', '-',"'"]

  for symbol in [".", ",", "!", "?", "-","'"]:
        issue = issue.replace(symbol, "")

  #issue = issue.replace('ok','okay')  #convert the slang

  # i got the replaced text of "ok" to "okay" repeatedly like "Okayayayay".
  # Here i used  to check the condition based - length of words

  words = issue.split()

  for i in range(len(words)):
    if words[i] == "ok":
        words[i] = "okay"

  issue = " ".join(words)

  issue = " ".join(issue.split())

  issue = issue.strip()

  cleaned_issues.append(issue)

ticket_data['Issue_Description'] = cleaned_issues

print("\nUpdated Tickets Data")
print('-' * 70)

for i in range(len(ticket_data['Ticket_No'])):

    print(f"Ticket No        : {ticket_data['Ticket_No'][i]}")
    print(f"Customer Name    : {ticket_data['Customer_Name'][i]}")
    print(f"Issue            : {ticket_data['Issue_Description'][i]}")
    print(f"Priority         : {ticket_data['Priority'][i]}")
    print("-" * 70)
```
